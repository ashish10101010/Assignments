```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Line Plot (sns.lineplot)
# Use: Displays data points connected by a line, useful for visualizing trends over time or continuous variables.

# Box Plot (sns.boxplot)
# Use: Shows the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum)
# useful for identifying outliers and comparing distributions across different categories.

# Histogram (sns.histplot)
# Use: Displays the distribution of a single numerical variable, useful for understanding the frequency and distribution of data points within intervals.

# Pair Plot (sns.pairplot)
# Use: Creates a matrix of scatter plots for multiple numerical variables, useful for exploring relationships between pairs of variables and understanding the overall structure of the data.

# Heatmap (sns.heatmap)
# Use: Displays data in a matrix form where values are represented by color intensity, useful for visualizing correlations between variables, frequency counts, or any matrix data.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')
plt.title('FMRI Signal over Time')
plt.show()
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'
# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Plot the box plots
plt.figure(figsize=(12, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')

plt.tight_layout()
plt.show()
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset
# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Price Distribution by Cut')
plt.show()
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset
# Load the iris dataset
iris = sns.load_dataset('iris')

# Plot the pair plot
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to get a matrix form
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Flights Dataset')
plt.show()
```
